```python
from sklearn.multiclass import OneVsRestClassifier
```

```python
clf.fit(x_train,y_train)
```

```python
f1_score(p3, y_test, average = 'weighted')
```

```python
from sklearn.ensemble import GradientBoostingClassifier
Classifier = GradientBoostingClassifier(n_estimators=100,max_depth=5)
Classifier.fit(x_train, y_train)
# make predictions for test data
p4 = Classifier.predict(x_test)
```

```python
f1_score(p4, y_test,average='weighted')
```

```python
from sklearn.neighbors import KNeighborsClassifier
Classifier = KNeighborsClassifier(n_neighbors=90)
Classifier.fit(x_train, y_train)
# make predictions for test data
p5 = Classifier.predict(x_test)
f1_score(p5, y_test,average='weighted')
```

```python
from sklearn import svm
svm_clf = svm.LinearSVC(random_state=14)
svm_clf.fit(x_train, y_train)
p6 = svm_clf.predict(x_test)
f1_score(p6, y_test,average='weighted')
```

```python
from xgboost.sklearn import XGBClassifier
Classifier = XGBClassifier()
Classifier.fit(x_train, y_train)
# make predictions for test data
p7 = Classifier.predict(x_test)
f1_score(p7, y_test,average='weighted')
```
